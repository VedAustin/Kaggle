```python
# First go to the folder ('D:/h2o) where use the command :java - jar h2o.jar
import h2o
h2o.init()
```

```python
df['TripType']=df['TripType'].asfactor()
```

```python
h2o.shutdown()
```

```python
# Generate random numbers and create training, validation and test splits
r = df.runif()
train = df[r < 0.8]
valid = df[(r >=0.8) & (r < 0.9)]
test = df[r>=0.9] 
```

```python
target = 'TripType'
features = ['Weekday','Upc','ScanCount','DepartmentDescription','FinelineNumber']
```

```python
modelGbm = h2o.gbm(y=target,x=features,training_frame=train,validation_frame=valid,ntrees=100,learn_rate=0.1,max_depth=4)
```

```python
modelGbm.params
```

```python
modelGbm
```

```python
prediction = modelGbm.predict(test)
pred = prediction.as_data_frame()
```
